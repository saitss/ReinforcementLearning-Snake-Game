```python
#Qlearning
pygame.init()
 
white = (255, 255, 255)
yellow = (255, 255, 102)
black = (0, 0, 0)
red = (213, 50, 80)
green = (0, 255, 0)
blue = (50, 153, 213)
grey=(128,128,128)
 
dis_width = 400
dis_height = 400
 
dis = pygame.display.set_mode((dis_width, dis_height))
clock = pygame.time.Clock()
 
snake_block = 40
snake_speed = 20
 
font_style = pygame.font.SysFont("bahnschrift", 25)
score_font = pygame.font.SysFont("comicsansms", 35)
 
 

 
 
    
def our_snake(state):
    for x in range(state.shape[0]):
        for y in range(state.shape[1]):
            if state[x,y]==2:
                pygame.draw.rect(dis, white, [x*snake_block, y*snake_block ,snake_block, snake_block])
            if state[x,y]==1:
                pygame.draw.rect(dis, grey, [x*snake_block, y*snake_block ,snake_block, snake_block])
            elif state[x,y]==5:
                pygame.draw.rect(dis, green, [x*snake_block, y*snake_block ,snake_block, snake_block])
 

 
def gameLoop():
    terminate=False
    game_over = False
    dis.fill(black)
    state=env.start()
    action=Qlearning.start(np.reshape(state,(10,10,1)))
    our_snake(state)
    pygame.display.update()
    while not game_over:
        reward,state,terminate=env.step(action)
        if(terminate):
            terminate=False
            action=Qlearning.end(reward)
            state=env.start()
            action=Qlearning.start(np.reshape(state,(10,10,1)))
        else:
            action=Qlearning.step(reward,np.reshape(state,(10,10,1)))
        dis.fill(black)
        our_snake(state)
        pygame.display.update()
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
        clock.tick(snake_speed)
 
    pygame.quit()
    
 
 
gameLoop()
```
